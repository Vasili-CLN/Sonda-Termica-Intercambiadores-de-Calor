```python
#Las Siguientes lineas no requieren de ninguna modificación, 
#son las librerias necesarias para el funcionamiento de la libreta
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from bokeh.plotting import figure, output_file, show
from bokeh.io import output_notebook
import matplotlib.colors as colors
output_notebook()
colores = list(colors._colors_full_map.values())
colores[1]= '#5698C6'
colores[2]= '#FF973B'
colores[3]= '#60B760'
colores[4]= '#E05C5D'
colores[5]= '#AC72A2'
colores[6]= '#8E594E'
colores[7]= '#FD9FD4'

archivo1='../Thingsboard/unidos.csv'

columnas2  = ['ts','A4','A5','A6','T3','T4']

def importa(archivo,columnas):
    
    tmp = pd.read_csv(archivo,header=None,sep=',',names = columnas,skiprows=1)
    #tmp = pd.read_csv(archivo,header=None,sep=',',names = columnas,skiprows=200)
    tmp=tmp.set_index('ts')
    tmp=tmp.dropna()
    return tmp
```

```python
datos=importa(archivo1,columnas2)
#datos.drop(['A3','T3','A5','A6'], axis=1, inplace=True)
datos.index=pd.to_datetime(datos.index,format='%Y-%m-%d  %H:%M:%S')
#datos.index=pd.to_datetime(datos.index,format='%dd/%mm/%YYYY  %H:%M:')

datos.plot()
```

```python
p = figure(title="Temperaturas subsuelo", x_axis_label='time', y_axis_label='Temperature [\xb0 C]',x_axis_type='datetime')
p.scatter(x=datos.index,y=datos.T3,color=colores[9],legend_label='Arduino_3')
p.scatter(x=datos.index,y=datos.A4,color=colores[17],legend_label='Adquisidor_4')
#p.scatter(x=datos.index,y=datos.A5,color=colores[18],legend_label='Adquisidor_5')
#p.scatter(x=datos.index,y=datos.A6,color=colores[20],legend_label='Adquisidor_6')
p.legend.location = "top_left"
show(p)
```
